```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.mixture import GaussianMixture
from sklearn.datasets import make_blobs
```

```python
df = pd.read_csv("Clustering_gmm.csv")
data_arr = df.to_numpy()
data_arr.shape
```

```python
# Plot the 2-D data
plt.scatter(data_arr[:,0], data_arr[:,1])
plt.show()
```

```python
gmm = GaussianMixture(n_components = 4)  
gmm.fit(data_arr)  
labels = gmm.predict(data_arr)

plt.scatter(data_arr[:,0], data_arr[:,1], c=labels)
plt.show()
```

```python
dummy_data, dummy_labels = make_blobs(n_samples=200, n_features=2, centers=4, shuffle=True, random_state=0)
pd.concat([pd.DataFrame(dummy_data), pd.DataFrame(dummy_labels)], axis=1)
```

```python
plt.scatter(dummy_data[:,0], dummy_data[:,1], c=dummy_labels)
plt.show()
```

```python
gmm.fit(dummy_data)  
gmm_labels = gmm.predict(dummy_data)

plt.scatter(dummy_data[:,0], dummy_data[:,1], c=gmm_labels)
plt.show()
```

```python
probs = gmm.predict_proba(dummy_data)
print(probs.round(3))
```
